```python
file = open("dataset_198_3.txt", mode="r")

patts = file.read().strip().splitlines()


def combine(string, patterns):
    for patt in patterns:
        if(string[len(string)-len(patt)+1:] == patt[:-1]):
            string += patt[-1]
    return string

def PathToGenome(patts):
    string = ""
    string += patts[0]
    Genome = combine(string, patts[1:])
    return Genome

print(PathToGenome(patts))
pTg = PathToGenome(patts)
txt = pTg
f = open('dataset_198_3.txt','w')
f.write(txt)
f.close()
```
